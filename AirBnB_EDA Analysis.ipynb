**Step 2: Data Cleaning**

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True) # Drop rows with any NaN values
data.isnull().sum() # Check for missing values again after dropping
```

```python
data.duplicated().sum() # Check for duplicate rows
```

```python
data.drop_duplicates(inplace=True) # Drop duplicate rows

# Check for duplicates again after dropping
data.duplicated().sum()
```

```python
data.rename(columns={'last_review': 'Last_Review_Date'}, inplace=True) # Rename column for clarity
data
```

```python
data['id'] = data['id'].astype(object) # Convert 'id' column to object type for consistency
data['host_id'] = data['host_id'].astype(object)
```

```python
data.info() # Check data types after conversion
```

**Step 3: Data Analysis**

```python
df = data[data['price'] < 1500]  # Filter out rows where price is greater than or equal to 1500
```

```python
df['neighbourhood_group'].value_counts() # Count the number of listings in each neighbourhood group
```

```python
top_neigh = df['neighbourhood'].value_counts().head(10) # Get the top 10 neighbourhoods by number of listings
top_neigh
```

```python
pivot = data.pivot_table(index='bedrooms', columns='room_type', values='price', aggfunc='mean')
pivot
```

**Step 4: Data visualization**

```python
#price distribution by frequency
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='price', bins=50)  
plt.title('Distribution of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
#Price Distribution by Room Type
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(x='room_type', y='price', data=df,width=0.5)
plt.title('Price Distribution by Room Type')
plt.ylim(0, 500)  # Remove outliers
plt.show()
```

```python
#Average Price by Neighbourhood Group
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='neighbourhood_group', y='price', data=df, estimator=np.mean)
plt.title('Average Price by Neighbourhood Group')
plt.ylim(0, 300)
plt.show()
```

```python
#Top Neighbourhoods by Number of Listings
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=top_neigh.values, y=top_neigh.index, palette='viridis')
plt.title('Top 10 Neighbourhoods by Listings')
plt.xlabel('Number of Listings')
plt.show()
```

```python
#Relationship Between Number of Reviews and Price
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='number_of_reviews', y='price', hue='room_type', alpha=0.5, data=df)
plt.ylim(0, 500)
plt.title('Price vs Number of Reviews by Room Type')
plt.show()
```

```python
#Availability vs. Rating
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='availability_365', y='rating', hue='room_type', data=df)
plt.xlabel('Availability (days per year)')
plt.ylabel('Rating')
plt.title('Availability vs Rating')
plt.legend(title='Room Type',bbox_to_anchor=(1.05, 1),loc='upper left')
plt.show()
```

```python
#Host Listing Count Analysis
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['calculated_host_listings_count'], bins=50, kde=True)
plt.title('Distribution of Host Listings Count')
plt.xlabel('Listings Count per Host')
plt.show()
```

```python
#Bedroom vs Price Heatmap
```

```python


plt.figure(figsize=(10,6))
sns.heatmap(pivot, annot=True, fmt=".0f", cmap='YlGnBu')
plt.xlabel("Room Type")
plt.ylabel("Number of Bedrooms")
plt.title('Average Price by Bedroom and Room Type')
plt.show()
```

```python
#Geographic Distribution
```

```python
sns.scatterplot(x='longitude', y='latitude', hue='neighbourhood_group', data=df, alpha=0.8)
plt.title('Geographic Distribution of Listings')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Neighbourhood Group', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
